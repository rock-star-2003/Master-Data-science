# Python Web Scraping
Web scraping is the process of extracting data from websites automatically. Python is widely used for web scraping because of its easy syntax and powerful libraries like BeautifulSoup, Scrapy, and Selenium.

## Installing Required Libraries
To install the required libraries in this article, run the following commands in the terminal.

 > - pip install requests
 > - pip install beautifulsoup4
 > - pip install selenium
 > - pip install lxml
 > - pip install schedule
 > - pip install pyautogui 

- <b>requests</b>: Sends HTTP requests to get webpage content (used for static sites).
- <b>beautifulsoup4</b>: Parses and extracts HTML content (like tags, text, links).
- <b>selenium</b>: Automates browsers (needed for dynamic sites with JavaScript).
- <b>lxml</b>: A fast HTML/XML parser, useful for large or complex pages.
- <b>schedule</b>: Lets you run scraping tasks repeatedly at fixed intervals.
- <b>pyautogui</b>: Automates mouse and keyboard; useful when dealing with UI-based interactions.

# Requests Module
The requests library is used for making HTTP requests to a specific URL and returns the response. Python requests provide inbuilt functionalities for managing both the request and response.

> - pip install requests

Example: Send a GET request to a webpage

```python
import requests

response = requests.get('https://www.geeksforgeeks.org/python/python-programming-language-tutorial/')

print(response.status_code)

print(response.content)
```

### Explanation:

- <b>requests.get(url)</b>: Sends a GET request to the given URL.
- <b>response.status_code</b>: Returns HTTP status code (200 = success).
- <b>response.content</b>: Returns the raw HTML of the page in bytes.
