### 1. What is RPA?
RPA, or **Robotic Process Automation**, is a technology that uses software robots to automate repetitive and rule-based business tasks. These bots can mimic human actions like logging into systems, reading emails, filling out forms, and updating records ‚Äî all without human intervention. It helps reduce manual effort and increases accuracy in business operations.

---

### 2. What are the applications of RPA?
RPA can automate many routine tasks, such as:
- Opening and reading emails  
- Extracting attachments  
- Moving or renaming files/folders  
- Filling in forms  
- Connecting to APIs  
- Reading/writing data to databases  
- Scraping data from websites or documents  
- Logging into enterprise systems to perform transactions  

---

### 3. What are the advantages of RPA?
Some key benefits include:
- **Cost reduction** ‚Äì up to 30% savings  
- **Higher accuracy** ‚Äì bots don‚Äôt make human errors  
- **Reduced operational risks** ‚Äì secure, auditable processes  
- **Scalability** ‚Äì easily add or remove bots  
- **Increased efficiency** ‚Äì employees focus on strategic work  
- **Improved compliance** and customer satisfaction  

---

### 4. Which industries benefit from RPA?
RPA is widely used in:
- **Banking & Finance** ‚Äì loan processing, reconciliations  
- **Insurance** ‚Äì claims handling  
- **Human Resources** ‚Äì onboarding, payroll  
- **Marketing & Sales** ‚Äì CRM data management, lead tracking

### 6. How does RPA work? (Lifecycle)
1. **Planning** ‚Äì Identify and analyze processes for automation  
2. **Design & Development** ‚Äì Build workflows using RPA tools  
3. **Deployment & Testing** ‚Äì Execute bots and validate results  
4. **Support & Maintenance** ‚Äì Monitor, update, and improve automation

### 7. Difference: Traditional Automation vs RPA
| Traditional Automation | RPA |
|-------------------------|-----|
| Requires API-level integration and system knowledge | Mimics user actions on the UI |
| Needs code-level changes | Works without modifying existing systems |

### 8. Types of RPA
- **Unattended Bots** ‚Äì Work in background, no human involvement  
- **Attended Bots** ‚Äì Triggered by the user, run on desktops  
- **Hybrid Bots** ‚Äì Combination of both

### 12. What is Citrix automation?
Citrix automation refers to automating tasks within virtual desktops (like Citrix environments). UiPath uses image recognition and OCR to handle this type of automation.

---

### 13. What are activities in UiPath?
Activities are the building blocks of automation in UiPath. Each activity performs a specific action such as:
- Reading/writing files  
- Handling data tables  
- Using logic and loops  
- Sending emails  
- Performing UI actions  

---

### 14. What is a sequence in UiPath?
A **sequence** is a simple linear workflow where activities are executed one after another. It‚Äôs used for small and straightforward processes.

---

### 15. What is scraping in UiPath?
UiPath provides two types of scraping:
- **Data Scraping** ‚Äì Extracts structured data (like tables)  
- **Screen Scraping** ‚Äì Captures text or images from the screen using OCR  

---

### 16. What is UiPath Studio?
UiPath Studio is the main development environment for creating automation workflows. It includes:
- Recorders (desktop, web, Citrix)  
- Debugging and logging tools  
- Reusable components  
- Visual workflow design  

---

### 17. What are UiPath Robots?
UiPath Robots execute the workflows built in UiPath Studio.
- **Attended Robots** ‚Äì Assist humans and are user-triggered  
- **Unattended Robots** ‚Äì Run autonomously in the background  

---

### 18. What does UiPath Orchestrator do?
UiPath Orchestrator is a centralized web-based platform used to deploy, manage, monitor, and schedule bots. It provides real-time logs, analytics, and user management.

---

### 19. Exception Handling in UiPath
Exception handling options include:
- **Try-Catch-Finally** ‚Äì Standard error handling block  
- **Throw** ‚Äì Raise custom exceptions  
- **Rethrow** ‚Äì Pass exception to parent workflow  
- **Terminate Workflow** ‚Äì Stop execution immediately  

---

### 20. What are Breakpoints in UiPath?
Breakpoints are used to pause workflow execution during debugging to inspect variable values and flow control.

---

### 21. What is a Macro Recorder?
A Macro Recorder records mouse and keyboard actions and converts them into automation steps ‚Äî useful for repetitive desktop tasks.

A Bot Runner is a system where bots are executed. It requires a ‚ÄúRun License‚Äù and is used strictly for running, not developing bots.

- **Task Bots** ‚Äì Handle repetitive, rule-based tasks

### 27. Is RPA same as screen scraping or macros?
No. RPA is much broader. Screen scraping and macros automate at a basic UI level, while RPA integrates with multiple systems, APIs, and databases, enabling end-to-end process automation.

| Type | Description |

| Chatbot | RPA |

| Can trigger RPA workflows | Completes tasks based on chatbot input |

### 30. What is Process Studio in RPA?
**Process Studio** is the workspace where automation logic is designed using loops, variables, and control flow elements. It‚Äôs similar to UiPath Studio but specific to tools like Blue Prism.
